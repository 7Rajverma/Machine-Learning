```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

# CATEGORICAL DATA

## A.Countplot

```python
# sns.countplot(df["Survived"])            # in this data 0 = dead and 1 = lived this both method work same 
# df["Survived"].value_counts()            # it show the data in the form of value 
df["Survived"].value_counts().plot(kind="bar")   #if we use .plot(kind="bar") then it show the bar graph 
```

```python
# sns.countplot(df["Pclass"])
df["Pclass"].value_counts().plot(kind="bar")
```

```python
# sns.countplot(df["Sex"])
df["Sex"].value_counts().plot(kind="bar")     # Smilarlly for other Categoricaldata
```

## B. Piechart

```python
df["Sex"].value_counts().plot(kind="pie") 
```

```python
df["Embarked"].value_counts().plot(kind="pie", autopct="%.2f")     # autopct="%.2f" show the percentage in the pie chart 
```
