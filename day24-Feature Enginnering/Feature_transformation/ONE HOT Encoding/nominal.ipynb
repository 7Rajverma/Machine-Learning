```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first", sparse_output=False,dtype=np.int32)     # Droping first columns 

X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_train_new

X_test_new =  ohe.transform(X_test[['fuel','owner']])
X_test_new
```

```python
repl = count[count<=threshold].index  # finding brand which has less then equal to 100  cars 
repl
```
