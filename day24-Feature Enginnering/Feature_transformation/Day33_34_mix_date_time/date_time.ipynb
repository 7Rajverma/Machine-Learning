```python
date['date'] = pd.to_datetime(date['date'])      #b conver the orject to datetime 
```

```python
date.info()
```

# 1.Extract Year

```python
date["date_year"] = date['date'].dt.year
date.head()
```

# 2.Extract Months

```python
date['date_months'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

# Extract Day

```python
date["date_day"] = date['date'].dt.day
date.head()
```

```python
# day week
date["date_day0fweek"] = date['date'].dt.dayofweek
date.head()
```

```python
# name_of a day
date["date_day_name"] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.week
date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract Semester

```python
date['date_semester'] =np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract Time elapsed between dates

```python
import datetime as dt

today = dt.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
