```python

from sklearn.datasets import make_regression      #this is use for creating dataset by its own 
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score     
```

```python
X,y = make_regression(n_samples=100,n_features=2,n_informative=2,n_targets=1, noise=50)  #n_sample=Row  , n_featurs=No. of Input_column, n_target=No.of output column
```

```python
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print('MAE', mean_absolute_error(y_pred, y_test))
print('MSE', mean_squared_error(y_pred, y_test))
print('R2 score', r2_score(y_pred, y_test))
```
