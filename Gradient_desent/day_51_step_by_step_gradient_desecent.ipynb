```python
from sklearn.datasets import make_regression
import numpy as np
```

# Finding m and b using LinearRegression (OLS) these are the correct value of slop and b and then comparing with the Gradient desent m and b

```python
X,y = make_regression(n_samples=4,n_features=1,n_targets=1,n_informative=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_      # value of m (slop) using linearRegression (OLS)
```

```python
reg.intercept_    # value of b using linearRegression (OLS)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')        
```

 ## lets apply the Gradient descent assuming slop is constant m = 78.35
## let as assume the statring value of intercept b = 0

```python
X
```

```python
y_pred = ((X * 78.35)+ 0).reshape(4)    # y value for  the line
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLS')     # using ols
plt.plot(X,y_pred,color='green', label='b=0')    
plt.legend()
plt.show()
```

## On above graph red line show the correct best fit line and green line is GD line the gole is to go as close as possible to the red line 

### Green line is GD line when we start GD when b = 0

```python
m = 78.35
b = 0

loss_slop = -2*np.sum(y - m*X.ravel()-b)
loss_slop
```

```python
# let learning rate 
l_rate = 0.1
step_size = l_rate*loss_slop
step_size
```

```python
# calculating b (intercept)
b = b - step_size        # New value of b
b
```

### We started with b = 0 and gole is to go at b=26.1596 now we are at the b= 20.927763

```python

y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='blue',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
     
```

```python
# Iteration 2
loss_slop = -2*np.sum(y - m*X.ravel()-b)
loss_slop
```

```python
step_size = l_rate*loss_slop
step_size
```

```python
b = b - step_size        # New value of b
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='blue',label='b=20.927763408209216')
plt.plot(X,y_pred2,color='yellow',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
     
```

### Now OLS is almost equal to GD value 26.15 == 26.16

```python
# Now using the loops
m = 78.35
b = -100
learning_rate = 0.01
epochs = 50
for i in range(epochs):
    loss_slop = -2*np.sum(y - m*X.ravel()-b)
    b = b - (learning_rate * loss_slop)

    y_pred = m * X+b
    plt.plot(X,y_pred)

plt.scatter(X,y)
    
```
